```python
import numpy as np
def llamarDatos (nombreArchivo):
    import os
    import pandas as pd
    serie=os.path.join('.',nombreArchivo+'.csv')
    serie_pd = pd.read_csv(serie, sep = ',',encoding='latin-1')
    return serie_pd

def Torque (nombreArchivo):
    import matplotlib.pyplot as plt
    data=llamarDatos(nombreArchivo)
    Torq=data['Torque (NÂ·m)'][~np.isnan(data['Torque (NÂ·m)'])]
    time=data['Time (s)'][~np.isnan(data['Torque (NÂ·m)'])]
    plt.plot(time,Torq)
    plt.xlabel('Time (s)')
    plt.ylabel('Torque [Nm]')
    return

def Thrust (nombreArchivo):
    import matplotlib.pyplot as plt
    data=llamarDatos(nombreArchivo)
    Thrust=data['Thrust (N)'][~np.isnan(data['Thrust (N)'])]
    time=data['Time (s)'][~np.isnan(data['Thrust (N)'])]
    plt.plot(time,Thrust,'b-')
    plt.xlabel('Time (s)')
    plt.ylabel('Thrust [N]')
    return
```

```python
x=llamarDatos('06-03-18(1)')
print(Torque('06-03-18(1)'))
```

```python
print(Thrust('06-03-18(1)'))
```
